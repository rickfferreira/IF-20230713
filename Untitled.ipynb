# AnÃ¡lise por consumo

```python
def consumo(x):
    dadosConsumox = dadosConsumo[['Class_Birth', 'Class_Income', 'Education', x, 'Kidhome', 'Teenhome', 'Marital_Status']]\
    .groupby(['Class_Birth', 'Class_Income', 'Education', 'Kidhome', 'Teenhome', 'Marital_Status'])\
    .describe().reset_index()
    dadosConsummoxClean = dadosConsumox.loc[dadosConsumox[(x, 'count')] >= 10]\
    .sort_values(by=(x, 'count'), ascending=False)\
    .drop([(x, 'min'), (x, '25%'),\
       (x, '50%'), (x, '75%'),\
       (x, 'max')], axis=1)
    return dadosConsummoxClean.sort_values(by=(x, 'mean'), ascending=False)
```

```python
def groupbyResponse1(x):
    dadosx = dadosMkt[['Class_Birth', 'Class_Income', 'Education', 'Kidhome', 'Teenhome', 'Marital_Status', 'Response', x]]\
    .groupby(['Class_Birth', 'Class_Income', 'Education', 'Kidhome', 'Teenhome', 'Marital_Status', 'Response'])\
    .describe().reset_index()
    dadosxClean = dadosx.loc[dadosx[(x, 'count')] >= 2]\
    .sort_values(by=(x, 'count'), ascending=False)\
    .drop([(x, 'min'), (x, '25%'),\
       (x, '50%'), (x, '75%'),\
       (x, 'max'), (x, 'std'), (x, 'mean')], axis=1)
    dadosFinal = dadosxClean.where(dadosxClean.Response == 1).sort_values(by=(x, 'count'), ascending=False)
    
    return dadosFinal.dropna()
```

```python
groupbyResponse1('MntWines').head()
```

```python
def groupbyResponse0(x):
    dadosx = dadosMkt[['Class_Birth', 'Class_Income', 'Education', 'Kidhome', 'Teenhome', 'Marital_Status', 'Response', x]]\
    .groupby(['Class_Birth', 'Class_Income', 'Education', 'Kidhome', 'Teenhome', 'Marital_Status', 'Response'])\
    .describe().reset_index()
    dadosxClean = dadosx.loc[dadosx[(x, 'count')] >= 2]\
    .sort_values(by=(x, 'count'), ascending=False)\
    .drop([(x, 'min'), (x, '25%'),\
       (x, '50%'), (x, '75%'),\
       (x, 'max'), (x, 'std'), (x, 'mean')], axis=1)
    dadosFinal = dadosxClean.where(dadosxClean.Response == 0).sort_values(by=(x, 'count'), ascending=False)
    
    return dadosFinal.dropna()
```
